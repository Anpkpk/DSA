```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern"],
    "font.size": 12
})

days = list(range(1, 16))
prices = [100, 113, 110, 85, 105, 102, 86, 63, 75, 95, 88, 92, 78, 105, 98]

plt.figure(figsize=(10, 5))
plt.plot(days, prices, marker='o', color='black', linestyle='-')
plt.grid(True, linestyle='--', alpha=0.5)
plt.ylim(50, 120)
plt.xticks(days)

plt.savefig("C:\\VSCode\\C++\\DSA\\lab3\\lab3_pdf\\plot.pdf")
plt.show()
```
