```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# D liu t kt qu chy
data = [
    (2, 0, 0),
    (4, 0, 0),
    (8, 0, 0),
    (16, 0, 0),
    (32, 0, 0),
    (64, 0, 0),
    (128, 0, 0),
    (256, 0, 0),
    (512, 0, 0),
    (1024, 0, 0),
    (2048, 0, 0),
    (4096, 0, 0),
    (8192, 0, 0),
    (16384, 0, 0),
    (32768, 0, 1.001),
    (65536, 0, 0.999),
    (131072, 0, 1.825),
    (262144, 0, 3.999),
    (524288, 0, 9.771)
]

# To DataFrame vi format yu cu
df = pd.DataFrame(data, columns=["Input Size", "Algorithm A (BinarySearch)", "Algorithm B (LinearSearch)"])

# Thm ct Ratio = B / A (x l A=0)
df["Ratio (B/A)"] = df.apply(lambda row: row["Algorithm B (LinearSearch)"]/row["Algorithm A (BinarySearch)"]
                             if row["Algorithm A (BinarySearch)"] > 0 else np.nan, axis=1)

# V biu  log-scale
plt.figure(figsize=(8,6))
plt.plot(df["Input Size"], df["Algorithm A (BinarySearch)"], marker='o', label="Binary Search")
plt.plot(df["Input Size"], df["Algorithm B (LinearSearch)"], marker='o', label="Linear Search")

plt.xscale("log", base=2)
plt.yscale("log")

plt.xlabel("Input Size (log scale)")
plt.ylabel("Execution Time (ms, log scale)")
plt.title("Growth Rate Comparison: Binary vs Linear Search")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.tight_layout()
plt.savefig("C:\\VSCode\\C++\\DSA\\lab2\\lab2_pdf\\growth_rate.pdf")
plt.show()


df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# D liu GCD
data = [
    (2, 0.0, 0.0),
    (4, 0.0, 0.0),
    (8, 0.0, 0.0),
    (16, 0.0, 0.0),
    (32, 0.0, 0.0),
    (64, 0.0, 0.0),
    (128, 0.0, 0.0),
    (256, 0.0, 0.0),
    (512, 0.0, 0.0),
    (1024, 0.0, 0.0),
    (2048, 0.0, 0.0),
    (4096, 0.0, 0.0),
    (8192, 0.0, 0.0),
    (16384, 0.0, 0.0),
    (32768, 0.0, 0.0),
    (65536, 0.0, 0.0),
    (131072, 0.0, 0.0),
    (262144, 0.0, 0.0),
    (524288, 0.0, 0.0),
]

# To DataFrame
df = pd.DataFrame(data, columns=["Input Size", "GCD Mod", "GCD Sub"])

# Thm ct Ratio (Sub / Mod)
df["Ratio"] = df.apply(lambda row: row["GCD Sub"]/row["GCD Mod"] if row["GCD Mod"] > 0 else None, axis=1)

# Xut CSV
df.to_csv("gcd_results.csv", index=False)

print(df)

# Xut LaTeX Table
latex_table = df.to_latex(index=False, float_format="%.4f", na_rep="--")
with open("gcd_table.tex", "w", encoding="utf-8") as f:
    f.write(latex_table)

# V biu  log-scale
plt.figure(figsize=(8,6))
plt.plot(df["Input Size"], df["GCD Mod"], marker='o', label="GCD (Modulo)")
plt.plot(df["Input Size"], df["GCD Sub"], marker='o', label="GCD (Subtraction)")

plt.xscale("log", base=2)
plt.yscale("log")

plt.xlabel("Input Size (log scale)")
plt.ylabel("Execution Time (ms, log scale)")
plt.title("Growth Rate Comparison: GCD Mod vs Subtraction")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.tight_layout()

plt.savefig("C:\\VSCode\\C++\\DSA\\lab2\\lab2_pdf\\gcd_comparison.pdf")
plt.show()
```
